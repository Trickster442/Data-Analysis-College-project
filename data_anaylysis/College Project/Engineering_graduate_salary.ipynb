# Working on Specialization column

```python
df['Specialization'].unique()
```

```python
df['Specialization'].value_counts()
```

```python
specialization = df['Specialization'].value_counts()
specialization
```

```python
lessthan20 = specialization[specialization <= 20]
lessthan20
```

# Assigning Name as others for values less than 20

```python
def renamingSpecialization(value):
    if value in lessthan20:
        return 'Others'
    else:
        return value

df['Specialization'] = df['Specialization'].apply(renamingSpecialization)
    
```

### Replacing -1 with Null values

```python
df = df.replace(-1 , np.nan)
df.head()
```

### Replacing Nan with men

```python
cols_with_nan = [column for column in df.columns if df.isna().sum()[column] > 0]
for column in cols_with_nan:
    df[column] = df[column].fillna(df[column].mean())
```
